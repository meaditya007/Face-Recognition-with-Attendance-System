```python
from tkinter import*
import tkinter as tk
from PIL import ImageTk,Image
from tkinter import messagebox
import tkinter.ttk as ttk

import mysql.connector
import cv2
import os
import numpy as np
from tkinter import PhotoImage

class Login:
    
    def __init__(self,root):
        self.root=root
        self.root.title("Login Page")
        self.root.geometry("1024x648+100+50")
        self.root.resizable(False,False)
        #bg image
        self.bg=PhotoImage(file="Login image1.png")
        self.bg_image=Label(self.root,image=self.bg).place(x=0,y=0,relwidth=1,relheight=1)
        
        #login frame
        frame_login = Frame(self.root,bg="#3090C7")
        frame_login.place(x=140,y=140,height=340,width=400)
        
        
        title = Label(frame_login,text="Login Here",font=("impact",35,"bold"),fg="#B7CEEC",bg="#3090C7").place(x=90,y=30)
        desc = Label(frame_login,text="Student Login Area",font=("Goudy old style",15,"bold"),fg="black",bg="#3090C7").place(x=90,y=100)
        
        lbl_user = Label(frame_login,text="UserName",font=("Goudy old style",15,"bold"),fg="black",bg="#3090C7").place(x=90,y=140)
        self.txt_user=Entry(frame_login,font=("times new roman",15),bg="light gray")
        self.txt_user.place(x=90,y=170,width=200,height=35)
        
        lbl_pass = Label(frame_login,text="Password",font=("Goudy old style",15,"bold"),fg="black",bg="#3090C7").place(x=90,y=210)
        self.txt_pass=Entry(frame_login,show="*",font=("times new roman",15),bg="light gray")
        self.txt_pass.place(x=90,y=240,width=200,height=35)
        
        #forget_btn = Button(frame_login,text="Forget Password?",cursor="hand2",bd=0,font=("times new roman",12),bg="#3090C7",fg="#9F000F").place(x=90,y=275)
        login_btn = Button(self.root,command=self.login,text="Login",cursor="hand2",font=("times new roman",15),fg="white",bg="#E77471").place(x=250,y=460,width=180,height=40)
    
    window=''
    def login(self):
        if self.txt_pass.get()== "" or self.txt_user.get()=="":
            messagebox.showerror("Error","All fields are required!!",parent=self.root)
        elif self.txt_pass.get()!= "Adi" or self.txt_user.get()!="Spider":
            messagebox.showerror("Error","Invalid Username/Password",parent=self.root)
        
        else:
            global window
            def another_window():
                #creating window
                
                window = tk.Toplevel()
                window.title('Face Recognition System')

                bg=PhotoImage(file="face2.png")
                bg_image=Label(window,image=bg).place(x=0,y=0,relwidth=1,relheight=1)

                #Registration frame
                registration_login = Frame(window,bg="#E0FFFF")
                registration_login.place(x=50,y=80,height=400,width=600)

                title = Label(registration_login,text="FACIAL RECOGNITION WINDOW",font=("snap itc",20),fg="black",bg="#E0FFFF").place(x=40,y=30)

                l1 = tk.Label(registration_login,text = " Name ",font=("Times New Roman",15)).place(x=60,y=100)
                t1=tk.Entry(registration_login,font=("times new roman",15),bg="white")             
                t1.place(x=200,y=100,width=200,height=25)

                l3 = tk.Label(registration_login,text = "Year",font=("Times New Roman",15)).place(x=60,y=140)              

                
                c1 = ttk.Combobox(registration_login)
                c1['values'] = ("1st","2nd","3rd","4th")
                c1.current(0)
                c1.place(x=200,y=140)

                l4 = tk.Label(registration_login,text = " Roll Number ",font=("Times New Roman",15)).place(x=60,y=180)           
                t4=tk.Entry(registration_login,font=("times new roman",15),bg="white")
                t4.place(x=200,y=180,width=200,height=25)

                l5 = tk.Label(registration_login,text = "Branch",font=("Times New Roman",15)).place(x=60,y=220)                
                
                c2 = ttk.Combobox(registration_login)
                c2['values'] = ("CSE","ECE","ME","CE","IT")
                c2.current(0)
                c2.place(x=200,y=220)

                l6 = tk.Label(registration_login,text = "Section",font=("Times New Roman",15)).place(x=60,y=260)               
                
                c3 = ttk.Combobox(registration_login)
                c3['values'] = ("A","B")
                c3.current(0)
                c3.place(x=200,y=260)


                def train_classifier():
                    data_dir = "C:/Users/Hp/Documents/Face Recognition/data"
                    path = [os.path.join(data_dir, f) for f in os.listdir(data_dir)]
                    faces = []
                    ids = []

                    for image in path:
                        img = Image.open(image).convert('L')  # another method to convert image to grayscale..
                        imageNp = np.array(img, 'uint8')
                        id = int(os.path.split(image)[1].split('.')[1])
                        faces.append(imageNp)
                        ids.append(id)
                    ids = np.array(ids)

                    # Train The Classifier and Save..
                    clf = cv2.face.LBPHFaceRecognizer_create()
                    clf.train(faces, ids)
                    clf.write('Classifier.xml')
                    messagebox.showinfo("Result","Training Dataset Completed!!!")

                b1 = tk.Button(window,text="Training",font=("snap itc",12),bg="light blue",fg="black",command=train_classifier)
                b1.place(x=60,y=435,width=180,height=40)

                def detect_face():
                    def draw_boundary(img, classifier, scaleFactor, minNeighbours, color, text, clf):
                        gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                        features = classifier.detectMultiScale(gray_img, scaleFactor, minNeighbours)

                        coords = []


                        for (x, y, w, h) in features:
                            cv2.rectangle(img, (x, y), (x + w, y + h), color, 2)
                            id, pred = clf.predict(gray_img[y:y + h, x:x + w])
                            confidence = int(100 * (1 - pred / 300))

                            mydb = mysql.connector.connect(
                            host="localhost",
                            user="root",
                            password="",
                            database="Authorised_user")
                            mycursor=mydb.cursor()
                            mycursor.execute("select name from my_table where id="+str(id))
                            s = mycursor.fetchone()
                            s = ""+"".join(s)
                            if confidence > 80:
                                cv2.putText(img, s, (x, y - 5), cv2.FONT_HERSHEY_COMPLEX, 0.8, color, 1, cv2.LINE_AA)

                            else:
                                cv2.putText(img, "Unknown", (x, y - 5), cv2.FONT_HERSHEY_COMPLEX, 0.8, color, 1, cv2.LINE_AA)
                            coords = [x, y, w, h]
                        return coords

                    def recognise(img, clf, faceCascade):
                        coords = draw_boundary(img, faceCascade, 1.1, 10, (255, 255, 255), "Face", clf)
                        return img

                    faceCascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
                    clf = cv2.face.LBPHFaceRecognizer_create()
                    clf.read("Classifier.xml")

                    video_capture = cv2.VideoCapture(0)

                    while True:
                        ret, img = video_capture.read()
                        img = recognise(img, clf, faceCascade)
                        cv2.imshow("FaceDetection", img)

                        if cv2.waitKey(1) == 13:
                            break

                    video_capture.release()
                    cv2.destroyAllWindows()


                b2 = tk.Button(window,text="Detect Face",font=("snap itc",12),bg="light blue",fg="black",command=detect_face)
                b2.place(x=260,y=435,width=180,height=40)

                def generate_dataset():
                    if (t1.get()=="" or c1.get()=="" or t4.get()=="" or c2.get()=="" or c3.get()==""):
                        messagebox.showinfo("Result","Please provide complete details of user!!")
                    else:
                        mydb = mysql.connector.connect(
                        host="localhost",
                        user="root",
                        password="",
                        database="Authorised_user")

                        mycursor=mydb.cursor()
                        mycursor.execute("SELECT * FROM my_table")
                        myresult = mycursor.fetchall()
                        id=1
                        for x in myresult:
                            id+=1
                        sql = "insert into my_table(id,Name,Year,ROllNumber,Branch,Section) values(%s,%s,%s,%s,%s,%s)"
                        val = (id,t1.get(),c1.get(),t4.get(),c2.get(),c3.get())
                        mycursor.execute(sql,val)
                        mydb.commit()

                        face_classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
                        def face_cropped(img):
                            gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
                            faces = face_classifier.detectMultiScale(gray,1.3,5)
                            if faces is ():
                                return None
                            for (x,y,w,h) in faces:
                                cropped_face = img[y:y+h,x:x+h]
                                return cropped_face
                        cap=cv2.VideoCapture(0)
                        id=1
                        img_id=0

                        while True:
                            ret,frame=cap.read()
                            if face_cropped(frame) is not None:
                                img_id+=1
                                face = cv2.resize(face_cropped(frame),(200,209))
                                face = cv2.cvtColor(face,cv2.COLOR_BGR2BGRA)
                                file_path_name = "data/user."+str(id)+"."+str(img_id)+".jpg"
                                cv2.imwrite(file_path_name,face)
                                cv2.putText(face,str(img_id),(50,50),cv2.FONT_HERSHEY_COMPLEX,1,(255,0,0),3)

                                cv2.imshow("Cropped Face",face)
                                if cv2.waitKey(1)==13 or int(img_id)==200:
                                    break
                        cap.release()
                        cv2.destroyAllWindows()
                        messagebox.showinfo('result','Generating Dataset Completed...')


                b3 = tk.Button(window,text="Generate Data Set",font=("snap itc",12),bg="light blue",fg="black",command=generate_dataset)
                b3.place(x=460,y=435,width=180,height=40)               

                window.geometry("1024x648+100+50")
                window.mainloop()

            another_window()
            
        

root=Tk()
obj=Login(root)

root.mainloop()
```
